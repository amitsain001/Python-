#### > View :

* The view() function is used to create a new view of the Series. This function returns a new Series with a view of the same underlying values in memory, optionally reinterpreted with a new data type. 

```python
newdef2= newdef        
newdef2[0][0]= 1234
newdef                           
```

* **newdef2 is a view of newdef not a copy of the newdef, it simply means that in memory newdef2 points to the newdef just like pointer and if we change the value in newdef2 then the value also changed in newdef and this is called as View.**

#### > Copy :

* The copy() method returns a copy of the DataFrame. By default, the copy is a "deep copy" meaning that any changes made in the original DataFrame will NOT be reflected in the copy.



```python
newdef2=newdef.copy()
newdef2[0][0]=3214
newdef
```

##### **NOTE : During setting up of a value we can't use copy and view function because it is not the guarantee of returing of a view or copy at a fix place after setting a value. It completely depends on the internal memory management of the pandas that what will be return either its view or its copy thats why we cant set value with the help of copy and view we use " loc() " to set the value.**

### **Merge DataFrames**

```python
df1= pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity":[68,65,75]
})
df2
```

```python
df3= pd.merge(df1,df2,on='city')
df3
```

**If two data frames have some different cities then during merging them takes intersection of the cities of two DataFrames.** 

```python
df1= pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["chicago","new york","san francisco"],
    "humidity":[65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

#### By default it takes intersection 👆 of the data of two DataFrames during Merging and display result which is also known as "**Inner**".

```python
df3=pd.merge(df1,df2,on='city',how="outer")
df3
```

### **Using "how=outer" above 👆 takes the union of the cities and displays the result after merging**

```python
df3=pd.merge(df1,df2,on='city',how='left')
df3
```

#### **here above 👆 "how='left' means that during merging it takes intersection of the data of two DataFrames and all data of df1 as well but not the rested data of the df2.**

```python
df3=pd.merge(df1,df2,on='city',how='right')
df3
```

#### **> indicator flag :**

* shows the DataFrame to which the NaN value belongs. 

```python
df3=pd.merge(df1,df2,on='city',how="outer", indicator=True)
df3
```
